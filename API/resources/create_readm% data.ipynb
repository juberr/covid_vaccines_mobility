```python
import pandas as pd
columns = ['race', 'gender', 'age', 'admission_type_id', 'discharge_disposition_id', 'admission_source_id', 'time_in_hospital', 'num_lab_procedures', 'num_procedures', 'num_medications', 'diag_1', 'diag_2', 'diag_3', 'number_diagnoses', 'max_glu_serum', 'a1cresult', 'diabetesmed', 'readmitted', 'num_meds_changed', 'total_hosp_visits']
df = pd.read_json('df_json.json')
```

```python
df.columns = columns
```

```python
df
```

```python
group_df = df.groupby(['diag_1']).sum()
```

```python
count_df = df.groupby(['diag_1']).count()
```

```python
group_df['count'] = count_df['gender']

group_df['readm%'] = group_df['readmitted']/group_df['count']
```

```python
final_df = group_df
```

```python
final_df.to_json('highestreadm.js', orient='values')
```
